# Convolutional Neural Networks

Convolutional Neural Networks (CNNs) are a type of neural network that have been very successful in image recognition and classification. One of the first, created in the late 80s was called LeNet. At the time it was mostly used for character recogntion tasks, such as the recognition of handwriting to help automate deliveries in the post office. Later in the 2010s CNNs resurfaced in popularity, this was largely because they performed extermemly well in an annual image recognition [competition](https://en.wikipedia.org/wiki/ImageNet#ImageNet_Challenge), in some cases outperforming humans. 


# Intro

Convolutional networks in additional to the normal fully connected layers also contain some additonal type of layers, namely convolutional layers and pooling layers. There are also some technical used terms such as padding, stride and kernel that we need to know.




## Convolution

What is a convolution? We can think of it intuitively as a type of filtering process. The convolution operation allows us to perform feature extraction, or determine the most important parts of an image. The figure bellow depicts a convolution.

![convultion example](http://deeplearning.net/software/theano/_images/numerical_no_padding_no_strides.gif)


The kernel (or filter) in the image above is the dark blue box sliding over the matrix .

$$ \quad \begin{pmatrix}
0 & 1 & 2 \\
2 & 2 & 0 \\
0 & 1 & 2
\end{pmatrix}
$$

We slide the kernel over the input and for each element that overlaps between the kernel and the input we calculate the product. We then sum the results to obtain the an output.
A good video explaning how [kernel](https://www.youtube.com/watch?v=C_zFhWdM4ic) are used in image processing

### Edge detection

The simplest feature we can extract is an edge, by using filters with particular values we can detect edges. For example the bellow kernel is a example of a verticle edge detectors.

 $$ \quad \begin{pmatrix}
1 & 0 & -1 \\
1 & 0 & -1 \\
1 & 0 & -1
\end{pmatrix}
$$

One commonly used edge detector is called the sobel filter. It is composed of two filters a horizonal one and a vertical one, shown bellow.






![Sobel filter](http://aishack.in/static/img/tut/conv-sobel.jpg)

Bellow is an example of appyling a sobel filter to a image, note how it sharpens the edges.


```python
from scipy import ndimage, misc
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure()
plt.gray()  # show the filtered result in grayscale
ax1 = fig.add_subplot(121)  # left side
ax2 = fig.add_subplot(122)  # right side
ascent = misc.ascent()
result = ndimage.sobel(ascent)
ax1.imshow(ascent)
ax2.imshow(result)
plt.show()
```



We can treat the kernels values as weights in our network, through training our network well learn there values. The kernels in earlier layers of the network will learn to detect simple features such as lines. In later layers they can combine these features detect more complex features like an edge or corner . As you move progressively deeper into the network it keeps combinining feature detctors eventually allowing it to classify objects like a boat or a house.


## Pooling 

Pooling helps to reduce the dimensionality of the data, reducing the number of weights, allowing us to train our models faster and prevent overfitting. [Max pooling](https://www.quora.com/What-is-max-pooling-in-convolutional-neural-networks) is commonly used which simply takes the greatest value from a group of numers.

![max pooling](https://ujwlkarn.files.wordpress.com/2016/08/screen-shot-2016-08-10-at-3-38-39-am.png?w=494)


## Padding

Padding is adding addtional zeros around the edge of a matrix. With convolution the images will shrink each convolution, so if we want the output matrix to be of the same size we have to pad the input matrix. In addtion without padding we don't use the pixels in the corners and edge of the image to there fullest, which is wastefull. 




![padding](http://deeplearning.net/software/theano/_images/same_padding_no_strides.gif)

## Stride

We can also slide the kernel by more than one step each time, the number of steps we move the kernel by is called the stride. Bellow is an example with a stride of 2.

![2 stride](http://deeplearning.net/software/theano/_images/no_padding_strides.gif)

# Convolutions with tensorflow

In the section well explore convlutions using tensorflow.

```python
from cv2 import imread #read image into an array
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import tensorflow as tf
%matplotlib inline
```

The image has a shape which describes it's height, width and 3 colour channels (RGB).

```python
img = imread("imgs/cat.jpeg")
img.shape
```

```python
plt.imshow(img)
plt.axis("off");
```

We can convert to grey scale by taking the mean across all of the colour channels.

```python
img = np.mean(img,axis=2)
plt.imshow(img,cmap="gray")
plt.axis("off");
```

In order to perform convlution in Tensorflow we need our image to be 4D, what each dimension means is shown bellow.


$$ N \times H \times W \times C = images $$

* $ N $ - number of images
* $ H $ - height of images
* $ W $ - width of images
* $ C $ - number of channels 

```python
img = img.astype("float32")
img_4d = tf.reshape(img,[1,img.shape[0],img.shape[1],1])
img_4d.shape.as_list()
```

Convlution kernel dimensitons

$$ K_H  \times K_W \times I  \times O $$

* $ K_H $ - kernel height
* $ K_W $ - kernel width
* $ I $ - number of input channels
* $ O $ - number of output channels


```python
kernel = np.array([[-1, 0, 1], 
                   [-1, 0, 1], 
                   [-1, 0, 1]]).astype("float32")


kernel_4d = tf.reshape(kernel,[3,3,1,1])
kernel_4d.shape.as_list()
```

```python
result = tf.nn.conv2d(img_4d,kernel_4d, strides=[1,1,1,1],padding="SAME")
with tf.Session() as sess:
    new_img = result.eval()
```

```python
plt.imshow(np.squeeze(new_img),cmap="gray")
plt.axis("off");
```

We could make this code more generic by using a placeholder, that way we can pass in an image of any shape.

```python
image_2d = tf.placeholder(tf.float32, shape=[None,None])
image_3d = tf.expand_dims(image_2d, 0)
image_4d = tf.expand_dims(image_3d ,3)
image_4d.shape.as_list()
```

```python
result= tf.nn.conv2d(image_4d, kernel_4d,
                          strides=[1, 1, 1, 1], padding='SAME')
```

```python
from skimage import data
```

```python
with tf.Session() as sess:
    img_one = sess.run(result,{image_2d:img})
    img_two = sess.run(result,{image_2d:data.camera()})
    
```

```python
fig, axs = plt.subplots(1,2)
axs[0].imshow(np.squeeze(img_one))
axs[1].imshow(np.squeeze(img_two))
```
